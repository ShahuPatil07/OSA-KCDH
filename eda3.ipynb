```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv("hour1_2.csv")
df1= df1.sort_values(by ="category")

df2 = pd.read_csv("hour2_2.csv")
df2= df2.sort_values(by ="category")

df3 = pd.read_csv("hour3_2.csv")
df3= df3.sort_values(by ="category")

df4 = pd.read_csv("hour4_2.csv")
df4= df4.sort_values(by ="category")

df5 = pd.read_csv("hour5_2.csv")
df5= df5.sort_values(by ="category")

df6 = pd.read_csv("hour6_2.csv")
df6= df6.sort_values(by ="category")

df7 = pd.read_csv("hour7_2.csv")
df7= df7.sort_values(by ="category")
```

```python
df7
```

```python
df = pd.DataFrame()
df["category"] = df1["category"].values
df["hour1_vc"] = df1["var_coeff"].values 
df["hour2_vc"] = df2["var_coeff"].values 
df["hour3_vc"] = df3["var_coeff"].values 
df["hour4_vc"] = df4["var_coeff"].values 
df["hour5_vc"] = df5["var_coeff"].values 
df["hour6_vc"] = df6["var_coeff"].values 
df["hour7_vc"] = df7["var_coeff"].values
df 
```

```python
len(df)
```

```python
fig, axes = plt.subplots(nrows=25, ncols=1, figsize=(10, 14), constrained_layout=True)

# Plot each category
for i in range(len(df)):
    hours = ['hour1_vc', 'hour2_vc', 'hour3_vc', 'hour4_vc', 'hour5_vc', 'hour6_vc', 'hour7_vc']
    values = df.loc[i, hours]
    axes[i].plot(range(1, 8), values, marker='o')
    axes[i].set_title(df.loc[i, 'category'])
    axes[i].set_xticks(range(1, 8))
    axes[i].set_xlabel('Hour')
    axes[i].set_ylabel('varying coefficient value')

# Show the plot
plt.show()
```

```python
for i in range(len(df)):
    hours = ['hour1_vc', 'hour2_vc', 'hour3_vc', 'hour4_vc', 'hour5_vc', 'hour6_vc', 'hour7_vc']
    values = df.loc[i, hours]
    plt.plot(range(1, 8), values, marker='o', label=df.loc[i, 'category'])

plt.xticks(range(1, 8))
plt.xlabel('Hour')
plt.ylabel('varying coefficient value')
plt.title('Value vs Hour for Different Categories')

```
