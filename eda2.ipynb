```python
df_mild = pd.read_csv("mild_stat3.csv")
df_moderate = pd.read_csv("moderate_stat3.csv")
df_norm = pd.read_csv("normal_stat3.csv")
df_severe = pd.read_csv("severe_stat3.csv")
```

```python
df = pd.DataFrame()
df["cat"] = cols
means1= []
for col in cols:
    means1.append(df_mild.loc[df_mild["cat"] == col, "mean"].values[0])
means2= []
for col in cols:
    means2.append(df_moderate.loc[df_moderate["cat"] == col, "mean"].values[0])   
means3= []
for col in cols:
    means3.append(df_norm.loc[df_norm["cat"] == col, "mean"].values[0]) 
means4= []
for col in cols:
    means4.append(df_severe.loc[df_severe["cat"] == col, "mean"].values[0])        
df["mild_mean"] = means1
df["moderate_mean"] = means2
df["normal_mean"] = means3
df["severe_mean"] = means4
```

```python
means=[]
for col in df["cat"].values:
    sum=0
    sum+=df.loc[df["cat"] == col, "mild_mean"].values[0]
    sum+=df.loc[df["cat"] == col, "moderate_mean"].values[0]
    sum+=df.loc[df["cat"] == col, "normal_mean"].values[0]
    sum+=df.loc[df["cat"] == col, "severe_mean"].values[0]
    means.append(sum/4)
std=[]    
for col in df["cat"].values:
    values=[df.loc[df["cat"] == col, "mild_mean"].values[0],
    df.loc[df["cat"] == col, "moderate_mean"].values[0],
    df.loc[df["cat"] == col, "normal_mean"].values[0],
    df.loc[df["cat"] == col, "severe_mean"].values[0]]
    std.append(np.array(values).std())
df["ovr_mean"] = means
df["std"] = std    

```

```python
df = df.dropna(subset=['var_coeff'])

percentile_1 = np.percentile(df['var_coeff'], 1)
percentile_99 = np.percentile(df['var_coeff'], 99)

# Step 2: Define the lower and upper thresholds
lower_threshold = 3 * percentile_1
upper_threshold = 3 * percentile_99

# Step 3: Filter the DataFrame to remove rows where 'data' is outside the thresholds
filtered_df = df[(df['var_coeff'] >= lower_threshold) & (df['var_coeff'] <= upper_threshold)]
```

```python
cats= filtered_df['category'].values
```

```python
import pyedflib

def read_edf(file_path):
    f = pyedflib.EdfReader(file_path)

    # Number of signals
    n = f.signals_in_file
    signal_labels = f.getSignalLabels()

    # Use a list to store each channel's data, since their lengths may differ
    sigbufs = []

    # Loop through each signal and read its data
    for i in np.arange(n):
        sigbufs.append(f.readSignal(i))

    # Close the EDF file
    f.close()

    # Convert the list of signals to a DataFrame (padded for unequal lengths)
    df_normal1 = pd.DataFrame({label: pd.Series(data) for label, data in zip(signal_labels, sigbufs)})
    return df_normal1
```

```python
file_path = r'Mild-20240823T063600Z-001\Mild\00000067-297469\00000067-297469[006].edf'
df_mild1= read_edf(file_path)[cats]

file_path = r'Moderate-20240823T064003Z-001\Moderate\00000074-297469\00000074-297469[006].edf'
df_moderate1 = read_edf(file_path)[cats]

file_path = r'Normal-20240823T064004Z-001\Normal\00000064-297469\00000064-297469[006].edf'
df_normal1 = read_edf(file_path)[cats]

file_path = r'severe-20240823T064005Z-001\severe\00000066-297469\00000066-297469[006].edf'
df_severe1= read_edf(file_path)[cats]
```

```python
means_mild = []
means_moderate= []
means_normal= []
means_severe = []
stat = pd.DataFrame()
for col in df_mild1.columns:
    means_mild.append(df_mild1[col].mean())
for col in df_moderate1.columns:
    means_moderate.append(df_moderate1[col].mean())
for col in df_normal1.columns:
    means_normal.append(df_normal1[col].mean()) 
for col in df_severe1.columns:
    means_severe.append(df_severe1[col].mean()) 
stat['category']= df_moderate1.columns
stat['mild']= means_mild
stat['moderate']= means_moderate              
stat['normal']= means_normal              
stat['severe']= means_severe  
stat["mean"] = (stat["mild"] + stat["moderate"] + stat["normal"] + stat["severe"])/4

std= []
for col in stat["category"].values:
    values=[stat.loc[stat["category"] == col, "mild"].values[0],
    stat.loc[stat["category"] == col, "moderate"].values[0],
    stat.loc[stat["category"] == col, "normal"].values[0],
    stat.loc[stat["category"] == col, "severe"].values[0]]
    std.append(np.array(values).std())      
stat["std"] = std
stat["var_coeff"] = abs(stat["std"]/stat["mean"])
stat = stat.sort_values(by = "var_coeff" ,ascending = False)    
stat    

```
