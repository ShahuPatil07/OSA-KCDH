```python
import pyedflib
import numpy as np
import pandas as pd

# Open the EDF file
def read_edf(file_path):
    f = pyedflib.EdfReader(file_path)

    # Number of signals
    n = f.signals_in_file
    signal_labels = f.getSignalLabels()

    # Use a list to store each channel's data, since their lengths may differ
    sigbufs = []

    # Loop through each signal and read its data
    for i in np.arange(n):
        sigbufs.append(f.readSignal(i))

    # Close the EDF file
    f.close()

    # Convert the list of signals to a DataFrame (padded for unequal lengths)
    df_normal1 = pd.DataFrame({label: pd.Series(data) for label, data in zip(signal_labels, sigbufs)})
    return df_normal1
```

```python
def normalize_column(df, column_name):
 
    df[column_name] = (df[column_name] - df[column_name].min()) / (df[column_name].max() - df[column_name].min())
    return df
```

```python
dfs = [df_normal1, df_normal2, df_normal3, df_normal4, df_normal5, df_normal6, df_normal7]
for df in dfs:
    for col in df.columns:
        df = normalize_column(df, col)
        print(str(col) + " done ")
```

```python

df_norm = pd.DataFrame()
df_norm['cat'] = df_normal1.columns.values
means = []
stds = []

for col in df_normal1.columns:
    sum1 = 0
    values1 = []

    for df in dfs:
        # Check for NaN values and extend only non-NaN values
        non_nan_values = df[col].dropna().tolist()  # Drop NaN values
        values1.extend(non_nan_values)              # Collect valid values

        # Calculate mean, ignoring NaN values
        sum1 += df[col].mean()

    # Compute mean across all DataFrames
    means.append(sum1 / len(dfs))

    # Convert to NumPy array and compute std, while handling empty arrays
    if values1:
        stds.append(np.array(values1).std())
    else:
        stds.append(np.nan)  # Append NaN if no values were collected

    print(col + " done")

# Create the DataFrame from means and stds
df_norm['mean'] = means
df_norm['std'] = stds
```
